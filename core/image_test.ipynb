```python
HEIGHT=256
WIDTH=256

trainPath = 'Train'
validPath = 'Valid'
testPath = 'Test'
```

```python
K.clear_session()
#Load weights
resnet = tf.keras.applications.resnet50.ResNet50(include_top=True, weights='imagenet', classes=1000)
```

```python
encoder_lab = ColorModel('encoder_lab')
encoder_color_lab = ColorModel('encoder_color_lab')
encoder_rgb = ColorModel('encoder_rgb')
encoder_color_rgb = ColorModel('encoder_color_rgb')
unet_rgb = ColorModel('unet_rgb')
unet_color_rgb = ColorModel('unet_color_rgb')
model_list = [encoder_lab, encoder_color_lab, encoder_rgb, encoder_color_rgb, unet_rgb, unet_color_rgb]
model_name = ['el', 'ecl', 'er', 'ecr', 'ur', 'ucr']
```

```python
# Output test for each of the models
for i in range(len(model_list)):
    model = model_list[i]
    name = model_name[i]
    model.pred_test(testPath, resnet, num_img=2, randomize=False, margin=60, is_save=name)
```

```python
# Testing color modes for colorizing page (only using the two best models)
model_best = [encoder_color_lab, unet_color_rgb]
name_best = ['ecl', 'ucr']
filename = 'pictures/sample_page_2.png'
color = [(255,0,0,1), (0,255,0,1), (55,200,255,1)]
for i in range(1):
    model = model_best[i]
    name = name_best[i]
    model.pred_page(filename, resnet, is_save=name)
    model.pred_page(filename, resnet, src_img='pictures/sample_3.jpg', is_save=name+'_src_img')
    model.pred_page(filename, resnet, src_color=color, is_save=name+'_src_color')
```
